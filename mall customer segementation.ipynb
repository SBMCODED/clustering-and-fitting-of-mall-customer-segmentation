#### line graph

For line plot: plt.plot(df[x_feature], df[y_feature], color='red', marker='o', linestyle='-', label='Line Plot'): This line plots a line plot using the specified x and y features from the DataFrame df. The line is represented in red color with circular markers and solid line style.

```python

def create_relational_graph(df, x_feature, y_feature, graph_type='scatter'):
    """
    Create a relational graph (scatter plot or line plot) to visualize the Mall Customer Segmentation Dataset.

    Parameters:
        df (DataFrame): The Mall Customer Segmentation Dataset.
        x_feature (str): The feature to plot on the x-axis.
        y_feature (str): The feature to plot on the y-axis.
        graph_type (str): Type of graph ('scatter' for scatter plot, 'line' for line plot). Default is 'scatter'.

    Returns:
        None. Displays the graph.
    """
    # Plot the graph
    plt.figure(figsize=(8, 6))
    if graph_type == 'scatter':
        plt.scatter(df[x_feature], df[y_feature], color='blue', marker='o', label='Data Points')
    elif graph_type == 'line':
        plt.plot(df[x_feature], df[y_feature], color='red', marker='o', linestyle='-', label='Line Plot')
    
    # Add labels and title
    plt.xlabel(x_feature)
    plt.ylabel(y_feature)
    plt.title(f'Relational Graph - {x_feature} vs {y_feature}')
    
    # Add grid for better readability
    plt.grid(True)
    
    # Add legend
    plt.legend()
    
    # Show plot
    plt.show()

# Example usage:
create_relational_graph(df, 'Annual Income (k$)', 'Spending Score (1-100)', graph_type='line')
```

#### Bar chart 
This code defines a function create_bar_chart that generates a bar chart to compare multiple categories from the Mall Customer Segmentation Dataset. Here's the breakdown:

def create_bar_chart(df, category_column, value_column, title, xlabel, ylabel):: This line defines the function create_bar_chart that takes six parameters:

df: The Mall Customer Segmentation Dataset, which is expected to be a DataFrame.
category_column: The column containing the categories to compare.
value_column: The column containing the values corresponding to each category.
title: The title of the bar chart.
xlabel: The label for the x-axis.
ylabel: The label for the y-axis.
grouped_data = df.groupby(category_column)[value_column].mean().reset_index(): This line groups the data by the category_column and calculates the mean values of value_column for each category. It then resets the index to make the result a DataFrame.

plt.figure(figsize=(10, 6)): This line creates a new figure with a specified figure size (10 inches wide and 6 inches tall).

plt.bar(grouped_data[category_column], grouped_data[value_column], color='skyblue'): This line plots a bar chart using the categories from category_column on the x-axis and the corresponding mean values from value_column on the y-axis. The bars are colored in sky blue.

plt.xlabel(xlabel) and plt.ylabel(ylabel): These lines label the x-axis and y-axis with the specified xlabel and ylabel.

plt.title(title): This line sets the title of the plot based on the specified title.

plt.xticks(rotation=45, ha='right'): This line rotates the x-axis labels by 45 degrees for better readability.

plt.show(): This line displays the plot.

The function doesn't return anything (None), but it displays the bar chart directly.

```python

def create_bar_chart(df, category_column, value_column, title, xlabel, ylabel):
    """
    Create a bar chart to compare multiple categories from the Mall Customer Segmentation Dataset.

    Parameters:
        df (DataFrame): The Mall Customer Segmentation Dataset.
        category_column (str): The column containing the categories to compare.
        value_column (str): The column containing the values corresponding to each category.
        title (str): The title of the bar chart.
        xlabel (str): The label for the x-axis.
        ylabel (str): The label for the y-axis.

    Returns:
        None. Displays the bar chart.
    """
    # Group data by category and calculate mean values
    grouped_data = df.groupby(category_column)[value_column].mean().reset_index()
    
    # Plot the bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(grouped_data[category_column], grouped_data[value_column], color='skyblue')
    
    # Add labels and title
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    
    # Rotate x-axis labels if needed for better readability
    plt.xticks(rotation=45, ha='right')
    
    # Show plot
    plt.show()

# Example usage:
# Assuming you want to compare average spending score by gender
category_column = 'Gender'
value_column = 'Spending Score (1-100)'
title = 'Average Spending Score by Gender'
xlabel = 'Gender'
ylabel = 'Average Spending Score'

create_bar_chart(df, category_column, value_column, title, xlabel, ylabel)
```

```python

def data_analysis(df):
    """
    Perform data analysis on the Mall Customer Segmentation Dataset.

    Parameters:
        df (DataFrame): The Mall Customer Segmentation Dataset.

    Returns:
        None. Displays analysis results and visualizations.
    """
    # Select numeric columns only
    numeric_columns = df.select_dtypes(include=['int64', 'float64'])
    
    # Perform statistical analysis
    descriptive_stats = numeric_columns.describe()  # Compute descriptive statistics
    correlation_matrix = numeric_columns.corr()     # Compute correlation matrix
    
    # Display descriptive statistics
    print("Descriptive Statistics:")
    print(descriptive_stats)
    print("\n")
    
    # Display correlation matrix
    print("Correlation Matrix:")
    print(correlation_matrix)
    print("\n")
    
    # Visualize data distribution using histograms
    numeric_columns.hist(figsize=(10, 8), bins=20)
    plt.suptitle('Histograms of Numerical Variables', x=0.5, y=1.02, fontsize=16)
    plt.tight_layout()
    plt.show()
    
    # Visualize correlation matrix using heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Matrix Heatmap')
    plt.show()



# Example usage:
data_analysis(df)
```

## statistical analysis 

The statistical_depth_analysis function conducts statistical depth analysis on the Mall Customer Segmentation Dataset. Here's a breakdown of its functionality:

Select Numeric Columns: It selects only the numeric columns from the DataFrame df.

Compute Major Moments: It computes major moments (mean, median, standard deviation, skewness, and kurtosis) for each numeric column and transposes the result for easier readability.

Compute Correlation Matrix: It computes the correlation matrix among the selected numeric columns.

Display Analysis Results:

Major Moments: It prints the major moments computed earlier.
Correlation Matrix: It prints the correlation matrix.
Basic Statistics: It prints basic statistics (count, mean, standard deviation, min, 25th percentile, median, 75th percentile, max) for the numeric columns.
Visualize Data Distribution:

Histograms: It visualizes the distribution of each numeric variable using histograms. The histograms are plotted with 20 bins and displayed in a grid layout.
Visualize Correlation Matrix:

Heatmap: It visualizes the correlation matrix using a heatmap. The heatmap displays the correlation coefficients between pairs of numeric variables. Higher positive correlations are represented by warmer colors (shades of red), while higher negative correlations are represented by cooler colors (shades of blue).
After defining the function, it loads the Mall Customer Segmentation Dataset from a CSV file (mall_customers.csv) into the DataFrame df, and then it calls the statistical_depth_analysis function with this DataFrame.

```python

def statistical_depth_analysis(df):
    """
    Perform statistical depth analysis on the Mall Customer Segmentation Dataset.

    Parameters:
        df (DataFrame): Mall Customer Segmentation Dataset DataFrame.

    Returns:
        None. Displays statistical depth analysis results and visualizations.
    """
    # Select numeric columns
    numeric_cols = df.select_dtypes(include=[np.number])
    
    # Compute major moments
    major_moments = numeric_cols.agg(['mean', 'median', 'std', 'skew', 'kurtosis']).transpose()
    print("Major Moments:")
    print(major_moments)
    print("\n")
    
    # Compute correlation matrix
    correlation_matrix = numeric_cols.corr()
    
    # Display correlation matrix
    print("Correlation Matrix:")
    print(correlation_matrix)
    print("\n")
    
    # Display basic statistics using describe
    print("Basic Statistics:")
    print(numeric_cols.describe())
    print("\n")
    
    # Visualize data distribution using histograms
    numeric_cols.hist(figsize=(10, 8), bins=20)
    plt.suptitle('Histograms of Numerical Variables', x=0.5, y=1.02, fontsize=16)
    plt.tight_layout()
    plt.show()
    
    # Visualize correlation matrix using heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Matrix Heatmap')
    plt.show()

# Perform statistical depth analysis on the Mall Customer Segmentation Dataset
statistical_depth_analysis(df)
```

## clustering function
This code defines a function perform_clustering that conducts K-means clustering on the Mall Customer Segmentation Dataset. Here's a breakdown of its functionality:

Select Relevant Features: It selects relevant features for clustering from the DataFrame df. In this case, it selects 'Annual Income (k$)' and 'Spending Score (1-100)' as features.

Initialize KMeans Model: It initializes a KMeans model with the specified number of clusters (num_clusters).

Perform Clustering: It fits the KMeans model to the selected features and assigns each data point in the DataFrame to a cluster based on the model's predictions. The cluster assignments are stored in a new column named 'Cluster' in the DataFrame df.

Return Clustered DataFrame: It returns the DataFrame df with an additional column indicating the cluster assignment for each data point.

After defining the function, it specifies the number of clusters (num_clusters) and calls the perform_clustering function with this parameter, resulting in a DataFrame (clustered_df) with cluster assignments for each data point. Finally, it prints the first few rows of the clustered DataFrame to examine the results.

The function returns the clustered DataFrame (df) with cluster assignments.

```python

from sklearn.cluster import KMeans
import warnings
# Filter out FutureWarnings from sklearn

def perform_clustering(df, num_clusters):
    """
    Perform K-means clustering on the Mall Customer Segmentation Dataset.

    Parameters:
        df (DataFrame): Mall Customer Segmentation Dataset DataFrame.
        num_clusters (int): Number of clusters to create.

    Returns:
        DataFrame: DataFrame with an additional column indicating the cluster assignment for each data point.
    """
    # Select relevant features for clustering (e.g., 'Annual Income' and 'Spending Score')
    X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

    # Initialize KMeans model with the specified number of clusters
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)

    # Perform clustering
    df['Cluster'] = kmeans.fit_predict(X)

    return df


# Specify the number of clusters
num_clusters = 5

# Perform clustering
clustered_df = perform_clustering(df, num_clusters)

# Display the first few rows of the clustered DataFrame
print(clustered_df.head())
```

## fitting function 
This code defines a function fit_line that fits a line to the data using linear regression and prints the slope and intercept of the fitted line. Here's a breakdown of its functionality:

Extract Variables: It extracts the independent variable (X) and the dependent variable (y) from the DataFrame df. The independent variable is specified by the x_column parameter, and the dependent variable is specified by the y_column parameter.

Initialize Linear Regression Model: It initializes a linear regression model (LinearRegression) from scikit-learn.

Fit Model to Data: It fits the linear regression model to the data using the independent variable X and the dependent variable y.

Get Slope and Intercept: It retrieves the slope and intercept of the fitted line from the trained linear regression model.

Print Slope and Intercept: It prints the slope and intercept of the fitted line.

After defining the function, it specifies the columns for the independent and dependent variables (x_column and y_column respectively) and calls the fit_line function with these parameters to fit the line to the data and print the result.

```python

from sklearn.linear_model import LinearRegression

def fit_line(df, x_column, y_column):
    """
    Fit a line to the data using linear regression and print the slope and intercept.

    Parameters:
        df (DataFrame): DataFrame containing the data.
        x_column (str): Name of the column containing the independent variable.
        y_column (str): Name of the column containing the dependent variable.

    Returns:
        None
    """
    # Extract independent and dependent variables
    X = df[[x_column]]
    y = df[y_column]

    # Initialize Linear Regression model
    model = LinearRegression()

    # Fit the model to the data
    model.fit(X, y)

    # Get the slope and intercept of the fitted line
    slope = model.coef_[0]
    intercept = model.intercept_

    # Print the slope and intercept
    print("Slope:", slope)
    print("Intercept:", intercept)


# Specify the columns for independent and dependent variables
x_column = 'Annual Income (k$)'
y_column = 'Spending Score (1-100)'

# Fit the line to the data and print the result
fit_line(df, x_column, y_column)
```

## clustering quality evaluation
This code performs clustering analysis on the Mall Customer Segmentation Dataset using K-means clustering. Here's a breakdown of the process:

Preprocess Data: The preprocess_data function is used to normalize the data and then scale it back to the original range. This is necessary for clustering algorithms like K-means, which are sensitive to the scale of the features.

Find Optimal Number of Clusters: The find_optimal_clusters function is employed to find the optimal number of clusters using two methods:

Silhouette Score: This metric measures how similar an object is to its own cluster compared to other clusters. Higher silhouette scores indicate better-defined clusters.
Elbow Method: This method looks for the "elbow" point in the plot of within-cluster sum of squares (WCSS) against the number of clusters. The elbow point represents the optimal number of clusters where adding more clusters does not significantly reduce the WCSS.
Visualize Optimal Number of Clusters: The silhouette scores and WCSS are plotted to visualize the optimal number of clusters based on the two methods.

Perform Clustering with Optimal Number of Clusters: K-means clustering is executed with the optimal number of clusters determined from the analysis.

Visualize Clustered Data: The clustered data points are visualized with colored groups representing different clusters, along with labeled cluster centers.

After running this code, you'll get visualizations indicating the optimal number of clusters and the clustered data points with cluster centers.

```python

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

def preprocess_data(df):
    """
    Preprocesses the data by normalizing it and then scaling it back to the original range.

    Args:
    - df (DataFrame): Input DataFrame containing the data.

    Returns:
    - scaled_data (array): Preprocessed data scaled back to the original range.
    """
    # Normalize the data
    scaler = StandardScaler()
    normalized_data = scaler.fit_transform(df)
    # Scale back to original range
    scaled_data = scaler.inverse_transform(normalized_data)
    return scaled_data

def find_optimal_clusters(data, max_clusters=10):
    """
    Finds the optimal number of clusters using silhouette score and elbow method.

    Args:
    - data (array): Input data for clustering.
    - max_clusters (int): Maximum number of clusters to consider.

    Returns:
    - silhouette_scores (list): Silhouette scores for each number of clusters.
    - wcss (list): Within-cluster sum of squares for each number of clusters.
    """
    silhouette_scores = []
    wcss = []
    for n_clusters in range(2, max_clusters + 1):
        kmeans = KMeans(n_clusters=n_clusters, random_state=42)
        kmeans.fit(data)
        cluster_labels = kmeans.labels_
        silhouette_avg = silhouette_score(data, cluster_labels)
        silhouette_scores.append(silhouette_avg)
        wcss.append(kmeans.inertia_)
    return silhouette_scores, wcss

def visualize_clusters(data, cluster_labels, cluster_centers):
    """
    Visualizes the clustered data with colored groups and labeled cluster centers in the legend.

    Args:
    - data (array): Input data.
    - cluster_labels (array): Labels of the clusters.
    - cluster_centers (array): Coordinates of the cluster centers.
    """
    plt.scatter(data[:, 0], data[:, 1], c=cluster_labels, cmap='viridis', alpha=0.5)
    plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', c='red', label='Cluster Centers')
    plt.legend()
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Clustered Data with Cluster Centers')
    plt.show()

# Load and preprocess the data (replace df with your data)
df = pd.read_csv("mall_customers.csv")
data = df[['Annual Income (k$)', 'Spending Score (1-100)']].values
scaled_data = preprocess_data(data)

# Find optimal number of clusters
silhouette_scores, wcss = find_optimal_clusters(scaled_data)

# Plot silhouette scores and WCSS
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(range(2, len(silhouette_scores) + 2), silhouette_scores, marker='o')
plt.title('Silhouette Score vs Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(range(2, len(wcss) + 2), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.tight_layout()
plt.show()

# Choose optimal number of clusters (e.g., by elbow method)
optimal_clusters = 5

# Perform clustering with optimal number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans.fit(scaled_data)
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

# Visualize clustered data
visualize_clusters(scaled_data, cluster_labels, cluster_centers)
```

## fitting quality evaluation 
This code defines two functions: fit_line and plot_fitted_line.

fit_line:

This function fits a line to the given data points using linear regression (linregress).
It takes two arguments: x (independent variable) and y (dependent variable).
It returns four values: slope (slope of the fitted line), intercept (intercept of the fitted line), slope_stderr (standard error of the slope), and intercept_stderr (standard error of the intercept).
plot_fitted_line:

This function plots the original data points along with the fitted line and confidence intervals.
It takes six arguments: x, y, slope, intercept, slope_stderr, and intercept_stderr.
It visualizes the original data points, the fitted line, and a 95% confidence interval around the fitted line.
The confidence interval is represented by shading around the fitted line.
After defining these functions, the code then extracts relevant columns for fitting (x and y), fits a line to the data using the fit_line function, and plots the fitted line with confidence intervals using the plot_fitted_line function.

```python
from scipy.stats import linregress

def fit_line(x, y):
    """
    Fits a line to the given data points.

    Args:
    - x (array-like): Independent variable.
    - y (array-like): Dependent variable.

    Returns:
    - slope (float): Slope of the fitted line.
    - intercept (float): Intercept of the fitted line.
    - slope_stderr (float): Standard error of the slope.
    - intercept_stderr (float): Standard error of the intercept.
    """
    slope, intercept, slope_stderr, intercept_stderr, _ = linregress(x, y)
    return slope, intercept, slope_stderr, intercept_stderr

def plot_fitted_line(x, y, slope, intercept, slope_stderr, intercept_stderr):
    """
    Plots the original data points along with the fitted line and confidence intervals.

    Args:
    - x (array-like): Independent variable.
    - y (array-like): Dependent variable.
    - slope (float): Slope of the fitted line.
    - intercept (float): Intercept of the fitted line.
    - slope_stderr (float): Standard error of the slope.
    - intercept_stderr (float): Standard error of the intercept.
    """
    plt.scatter(x, y, label='Original Data')
    plt.plot(x, slope * x + intercept, color='red', label='Fitted Line')
    plt.fill_between(x, (slope - slope_stderr) * x + (intercept - intercept_stderr), (slope + slope_stderr) * x + (intercept + intercept_stderr), color='gray', alpha=0.2, label='95% Confidence Interval')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Fitted Line with Confidence Interval')
    plt.legend()
    plt.grid(True)
    plt.show()

# Extract relevant columns for fitting (replace 'actual_column_name1' and 'actual_column_name2' with actual column names)
x = df['Age'].values
y = df['Annual Income (k$)'].values

# Fit a line to the data
slope, intercept, slope_stderr, intercept_stderr = fit_line(x, y)

# Plot the fitted line with confidence intervals
plot_fitted_line(x, y, slope, intercept, slope_stderr, intercept_stderr)
```

## fitting prediction/ accuracy of fitting prediction 
The provided code defines two functions: fit_line and plot_fitted_line_with_uncertainty.

fit_line:

This function fits a line to the given data points using linear regression (linregress).
It takes two arguments: x (independent variable) and y (dependent variable).
It returns four values: slope (slope of the fitted line), intercept (intercept of the fitted line), slope_stderr (standard error of the slope), and intercept_stderr (standard error of the intercept).
plot_fitted_line_with_uncertainty:

This function plots the original data points along with the fitted line and uncertainty predictions.
It takes six arguments: x, y, slope, intercept, slope_stderr, and intercept_stderr.
It also takes an optional argument num_predictions to specify the number of uncertainty predictions to generate and plot.
Uncertainty predictions are generated by adding random uncertainties to the fitted line based on the standard errors of the slope and intercept.
Each uncertainty prediction is plotted as a gray line with low transparency (alpha=0.3) to represent the range of possible outcomes.
After defining these functions, the code loads the dataset from the file "mall_customers.csv" into a DataFrame df, extracts relevant columns ('Age' and 'Annual Income (k$)') for fitting, fits a line to the data using the fit_line function, and plots the fitted line with uncertainty predictions using the plot_fitted_line_with_uncertainty function.

```python
from scipy.stats import linregress

def fit_line(x, y):
    """
    Fits a line to the given data points and calculates uncertainties.

    Args:
    - x (array-like): Independent variable.
    - y (array-like): Dependent variable.

    Returns:
    - slope (float): Slope of the fitted line.
    - intercept (float): Intercept of the fitted line.
    - slope_stderr (float): Standard error of the slope.
    - intercept_stderr (float): Standard error of the intercept.
    """
    slope, intercept, slope_stderr, intercept_stderr, _ = linregress(x, y)
    return slope, intercept, slope_stderr, intercept_stderr

def plot_fitted_line_with_uncertainty(x, y, slope, intercept, slope_stderr, intercept_stderr, num_predictions=20):
    """
    Plots the original data points along with the fitted line and uncertainty predictions.

    Args:
    - x (array-like): Independent variable.
    - y (array-like): Dependent variable.
    - slope (float): Slope of the fitted line.
    - intercept (float): Intercept of the fitted line.
    - slope_stderr (float): Standard error of the slope.
    - intercept_stderr (float): Standard error of the intercept.
    - num_predictions (int): Number of predictions to generate.

    Returns:
    - None
    """
    plt.scatter(x, y, label='Original Data')
    plt.plot(x, slope * x + intercept, color='red', label='Fitted Line')
    
    for _ in range(num_predictions):
        # Generate random uncertainties based on standard errors, ensuring scale parameter is non-negative
        slope_uncertainty = max(0, slope_stderr) * np.random.randn(len(x))
        intercept_uncertainty = max(0, intercept_stderr) * np.random.randn(len(x))
        
        # Generate prediction with uncertainties
        prediction = slope * x + intercept + slope_uncertainty * x + intercept_uncertainty
        
        # Plot prediction with uncertainties
        plt.plot(x, prediction, color='gray', alpha=0.3)

    plt.xlabel('Age')
    plt.ylabel('Annual Income (k$)')
    plt.title('Fitted Line with Uncertainty Predictions')
    plt.legend()
    plt.grid(True)
    plt.show()

# Load data (replace 'mall_customers.csv' with the actual filename of your data)
df = pd.read_csv("mall_customers.csv")

# Extract relevant columns for fitting
x = df['Age'].values
y = df['Annual Income (k$)'].values

# Fit a line to the data
slope, intercept, slope_stderr, intercept_stderr = fit_line(x, y)

# Plot the fitted line with uncertainty predictions
plot_fitted_line_with_uncertainty(x, y, slope, intercept, slope_stderr, intercept_stderr)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit

def load_data(file_path):
    """
    Load the Mall Segmentation dataset from the specified file path.

    Args:
    - file_path (str): Path to the dataset file.

    Returns:
    - df (DataFrame): Pandas DataFrame containing the dataset.
    """
    return pd.read_csv(file_path)

def preprocess_data(df):
    """
    Preprocess the dataset by handling missing values and encoding categorical variables if needed.

    Args:
    - df (DataFrame): Input DataFrame.

    Returns:
    - X (array-like): Features (independent variables).
    - y (array-like): Target variable (dependent variable).
    """
    X = df[['Age']].values
    y = df['Annual Income (k$)'].values
    return X, y

def polynomial_fitting(X, y, degree=2):
    """
    Perform polynomial fitting on the data.

    Args:
    - X (array-like): Features (independent variables).
    - y (array-like): Target variable (dependent variable).
    - degree (int): Degree of the polynomial.

    Returns:
    - model (LinearRegression): Fitted polynomial regression model.
    """
    poly_features = PolynomialFeatures(degree=degree)
    X_poly = poly_features.fit_transform(X)
    model = LinearRegression()
    model.fit(X_poly, y)
    return model

def exponential_fitting(X, y):
    """
    Perform exponential fitting on the data.

    Args:
    - X (array-like): Features (independent variables).
    - y (array-like): Target variable (dependent variable).

    Returns:
    - params (array-like): Fitted parameters of the exponential function.
    """
    def exponential_func(x, a, b, c):
        return a * np.exp(b * x) + c
    params, _ = curve_fit(exponential_func, X.flatten(), y)
    return params


# Preprocess data
X, y = preprocess_data(df)

# Perform polynomial fitting
degree = 2
poly_model = polynomial_fitting(X, y, degree)
X_poly = PolynomialFeatures(degree=degree).fit_transform(X)

# Perform exponential fitting
exp_params = exponential_fitting(X, y)

# Plotting
plt.figure(figsize=(12, 6))

# Scatter plot of original data
plt.scatter(X, y, color='blue', label='Original Data')

# Plot polynomial fit
X_plot = np.linspace(min(X), max(X), 100).reshape(-1, 1)
plt.plot(X_plot, poly_model.predict(PolynomialFeatures(degree=degree).fit_transform(X_plot)), color='red', label='Polynomial Fit')

# Plot exponential fit
def exponential_func(x, a, b, c):
    return a * np.exp(b * x) + c
plt.plot(X_plot, exponential_func(X_plot, *exp_params), color='green', label='Exponential Fit')

plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('Polynomial and Exponential Fitting')
plt.legend()
plt.grid(True)
plt.show()
```
